```python
import requests
```

```python
g = []
def from_json(json_dict, prefix=''):
    for k, v in json_dict.items():
        if isinstance(v, dict):
            p = "{}.{}".format(prefix, k)
            from_json(v, p)
        elif isinstance(v, list):
            for i, v2 in enumerate(v):
                p = "{}{}".format(prefix, i)
                from_json(v2, p)
        else:
            g.append(['{}'.format(prefix),v])
    return g
```
