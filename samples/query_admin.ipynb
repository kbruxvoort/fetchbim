```python
from fetchbim.family import Family
from fetchbim.query import Filter
from fetchbim.admin import get_family
from fetchbim.settings import DELETE_FAMILY
```

```python
penco_filter = Filter(ParameterName='zM', ParameterValue='33072', ParameterValueMatchType=3)
families = penco_filter.query()
```

```python
for fam_dict in families:
    fam = Family.from_json(fam_dict)
    print(repr(fam))
    # print(fam.delete().json())
```

```python
get_family(fam.Id)
```

```python
for fam_dict in families:
    fam = Family.from_json(fam_dict)
    fam.restore()
```

```python
get_family(fam.Id)
```
