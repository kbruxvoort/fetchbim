```python
from fetchbim.family import Family
from fetchbim.query import Filter
from fetchbim.admin import get_family

from re import sub
from decimal import Decimal
```

```python
ham_filter = Filter(CategoryName="Model Groups/Blood Slide Vertical Carousels")
families = ham_filter.query()
# print(families)
```

```python
for fam_dict in families[:1]:
    fam = Family.from_json(fam_dict)
    full_fam = Family.from_json(get_family(fam.Id))
    full_fam.Files = []
    full_fam.Parameters = []
    print(full_fam.Name)
    for prop in full_fam.Properties:
        if prop.Name == "Budget Price":
            old_value = Decimal(sub(r'[^\d.]', '', prop.Value))
            new_value = round(old_value * Decimal(1.244), 2)
            print(f'\t{prop.Name}: {old_value} → {new_value}')
            prop.Value = str(new_value)
            # full_fam.post()
            print(repr(full_fam))

    # print(fam.Id)
    # print(fam.delete().json())
```
